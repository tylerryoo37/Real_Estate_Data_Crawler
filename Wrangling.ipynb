### CHECK COLUMNS BEFORE PROCEEDING 

```python
df = pd.concat([df, df['new_info'].apply(pd.Series)], axis=1)
```

```python
df = df.drop(columns = ['unit_info'])
```

```python
final_result = df.drop(columns = ['new_info'])
```

```python
final_result.head()
```

```python
df["chambres/pièces_fixed"] = df['chambres/pièces'].add(df['chambre/pièce'], fill_value = "") 
```

```python
df["bains/douches_fixed"] = df['bains/douches'].add(df['bain/douche'], fill_value = "") 
```

```python
# df["salles de bain_fixed"] = df['salles de bain'].add(df['salle de bain'], fill_value = "") 
```

```python
df.head()
```

```python
result = pd.DataFrame(
            {"Property": df['property'], 
             "Price_Range": df['price_range'],
             "Location": df['location'],
             "Link": df['property_link'],          
             "Agency": df['agency'],
             "Unit": df['unit'],
             "Unit_Price_euro": df['unit_price'],
             "Unit_Type" : df['unit_type'],
             "Rooms":df['chambres/pièces_fixed'],
             "Area_sqm":df['superficie'],
             "Bathrooms":df['bains/douches_fixed'],
             "Stage":df['étage']})
```

```python
result
```

```python
unit_price = []
for i in result['Unit_Price_euro']:
    if pd.isna(i) or i == "Prix sur demande":
        unit_price.append(None)
    else: 
        unit_price.append(int(str(i).replace(",","").lstrip('€ ')))
```

```python
result['Unit_Price_euro'] = unit_price
```

```python
result['Unit_Price_euro'] = pd.to_numeric(result['Unit_Price_euro'])
```

```python
area = []
for i in result['Area_sqm']:
    if pd.isna(i):
        area.append(None)
    else: 
        area.append(int(str(i).replace(".","").rstrip('\n m²')))
```

```python
result['Area_sqm'] = area
```

```python
result['Area_sqm'] = pd.to_numeric(result['Area_sqm'])
```

```python
rooms = []
for i in result['Rooms']:
    if pd.isna(i):
        rooms.append(None)
    else: 
        rooms.append(int(str(i).replace(".","").rstrip('+')))
```

```python
result['Rooms'] = rooms
```

```python
result['Rooms'] = pd.to_numeric(result['Rooms'])
```

```python
bathroom = []
for i in result['Bathrooms']:
    if pd.isna(i):
        bathroom.append(None)
    else: 
        bathroom.append(int(str(i).replace(".","").rstrip('+')))
```

```python
result['Bathrooms'] = bathroom
```

```python
result['Bathrooms'] = pd.to_numeric(result['Bathrooms'])
```

```python
stage = []
for i in result['Stage']:
    if pd.isna(i):
        stage.append(pd.to_numeric(None))
    elif i == 'T': 
        stage.append('T')
    elif i == 'R':
        stage.append('R')
    else: 
        stage.append(int(i))
```

```python
result['Stage'] = stage
```

```python
result
```

### Change File Name Before Running

```python
result.to_csv(sales_file_name_csv, encoding = 'utf-8-sig')
```

```python
result.to_excel(sales_file_name_xlsx, encoding = 'utf-8-sig')  
```
